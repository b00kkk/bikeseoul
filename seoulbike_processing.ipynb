서울특별시 공공자전거 이용정보(시간대별) 전처리

```python
usage=pd.read_csv('/content/서울특별시 공공자전거 이용정보(시간대별)_2312.csv',encoding='euc-kr')
```

```python
print(usage.isna().sum())
```

```python
usage=usage.drop(columns='성별')
```

```python
def station_name(text):
    parts = text.split('.', 1)
    if len(parts) > 1:
        return parts[1].strip()
    else:
        return text.strip()

usage['대여소명'] = usage['대여소명'].apply(station_name)
usage.head()
```

```python
usage.to_csv('hour_usage.csv',index=False)
pd.read_csv('hour_usage.csv')
```
